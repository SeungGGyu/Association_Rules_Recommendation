```python
import pandas as pd

# 파일 경로
file_path = r'C:\Users\Administrator\Desktop\대학\3학년 1학기\데이터마이닝 입문\Processed_file_final_name.csv'
data = pd.read_csv(file_path, encoding='utf-8')
data.head()
```

```python
# taste_score와 amount_score가 모두 4 이상인 데이터 필터링
filtered_data = data[(data['taste_score'] >= 4) & (data['amount_score'] >= 4)]
filtered_data
```

```python
# 필터링된 데이터를 새로운 CSV 파일로 저장
output_file_path = 'C:/Users/Administrator/Desktop/대학/3학년 1학기/데이터마이닝 입문/filtered_data.csv'
filtered_data.to_csv(output_file_path, index=False, encoding='utf-8-sig')
```
